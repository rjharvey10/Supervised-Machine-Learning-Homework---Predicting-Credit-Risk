```python
import numpy as np
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, LabelEncoder
import seaborn as sns
```

```python
train_df = pd.read_csv(Path('Resources/2019loans.csv'))
test_df = pd.read_csv(Path('Resources/2020Q1loans.csv'))
train_df.head(10)
```

My prediction for the more accurate model, when comparing Logistical Regression and Random Forest, is RF because it has many more factors and variables which allow RF to be more maleable and react to the model rather than having forced linear elements which can skew the model. RF can also make decisions in it's Decision Tree to react, while LR does not have the capability.

```python
# Convert categorical data to numeric and separate loan_status feature for training data
y_train=train_df["loan_status"]
X_train= train_df.drop(columns=["loan_status"])
X_train.head(10)
```

```python
# Convert categorical data to numeric and separate loan_status feature for testing data
y_test=test_df["loan_status"]
X_test= test_df.drop(columns=["loan_status"])
X_test.head(10)
```

```python
# add missing dummy variables to testing set
X_dummies_train = pd.get_dummies(X_train)
X_dummies_test = pd.get_dummies(X_test)
print(f"Train: {X_dummies_train.shape}, Test: {X_dummies_test.shape}")
```

```python
y_label_train = LabelEncoder().fit_transform(train_df['loan_status'])
y_label_train

y_label_test = LabelEncoder().fit_transform(test_df['loan_status'])
y_label_test
```

```python
for col in X_dummies_train.columns:
    if col not in X_dummies_test.columns:
        X_dummies_test[col]=0
```

```python
# Train the Logistic Regression model on the unscaled data and print the model score
from sklearn.linear_model import LogisticRegression
classifier_logistic  = LogisticRegression(solver='lbfgs').fit(X_dummies_train, y_label_train) 

print(f"Training Model Score: {classifier_logistic.score(X_dummies_train, y_label_train)}")
print(f"Testing Model Score: {classifier_logistic.score(X_dummies_test, y_label_test)}");
```

```python
# Train a Random Forest Classifier model and print the model score
from sklearn.ensemble import RandomForestClassifier
classifier_forest = RandomForestClassifier(random_state=42, n_estimators=200).fit(X_dummies_train, y_label_train)

print(f'Training Model Score: {classifier_forest.score(X_dummies_train, y_label_train)}')
print(f'Testing Model Score: {classifier_forest.score(X_dummies_test, y_label_test)}')
```

Prediction Result: It appears my prediction was correct (at least for unscaled data), as both the Training Model and Testing Model Scores were higher for Random Forest. The results of the two models are below:

```python
# Scale the data
scaler = StandardScaler().fit(X_dummies_train)
X_train_scaled = scaler.transform(X_dummies_train)
X_test_scaled = scaler.transform(X_dummies_test)
```

```python
# Train the Logistic Regression model on the scaled data and print the model score
classifier_lr = LogisticRegression().fit(X_train_scaled, y_label_train)

print(f'Training Model Score: {classifier_lr.score(X_test_scaled, y_label_test)}')
```

```python
# Train a Random Forest Classifier model on the scaled data and print the model score
classifier_rf = RandomForestClassifier(random_state=42, n_estimators=500).fit(X_train_scaled, y_label_train)

print(f'Testing  Model Score: {classifier_rf.score(X_test_scaled, y_label_test)}')
```

Interestingly, the Logistical Regression Model is the more accurate model after scaling has been applied to the data. The LR model may be more effective at scaling and predicting a more linear set of data in an open set, while RF is better at predicting more variable data in an already established set of data.
